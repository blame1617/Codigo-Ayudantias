### 📦 Introducción a Librerías en Python

Una librería es un conjunto de herramientas y funciones preescritas que podemos importar a nuestro programa para no tener que inventar la rueda desde cero. Hoy veremos las tres más esenciales: math (matemáticas), random (aleatoriedad) y time (manejo del tiempo).

El primer paso es importar las librerías al archivo en el que estamos trabajando. Se debe hacer en las primeras líneas.

```python
# Importamos las librerías al inicio de nuestro notebook
import math
import random
import time
```

### 1. Librería math

Esta librería nos da acceso a operaciones matemáticas avanzadas que van mucho más allá de sumar, restar, multiplicar o dividir.

Funciones clave:

    math.ceil(x): Redondea un número decimal siempre hacia arriba al entero más cercano.

    math.floor(x): Redondea un número decimal siempre hacia abajo al entero más cercano.

    math.sqrt(x): Calcula la raíz cuadrada exacta.

    math.pi: Nos entrega el valor de Pi (π) con alta precisión.

```python
print("--- APLICACIONES DE MATH ---\n")

# Ejemplo 1: Comprando equipamiento (Uso de ceil)
metros_necesarios = 245.5
metros_por_bobina = 100
bobinas_a_comprar = metros_necesarios / metros_por_bobina # Da 2.455

# No podemos comprar 2.455 bobinas de hilo, necesitamos redondear hacia arriba
bobinas_reales = math.ceil(bobinas_a_comprar)
print(f"1. Necesitamos comprar {bobinas_reales} bobinas de hilo (Redondeo hacia arriba).")

# Ejemplo 2: Cálculo de raíces (Uso de sqrt)
distancia_x = 9
distancia_y = 16
distancia_real = math.sqrt(distancia_x + distancia_y) 
print(f"2. La distancia exacta entre los puntos es: {distancia_real}")

# Ejemplo 3: Cálculo circular (Uso de pi)
radio = 5
area_circulo = math.pi * (radio ** 2)
print(f"3. El área del circulo es: {area_circulo} m2.")
```

### 2. Librería random 

Fundamental para crear simulaciones, videojuegos, o cualquier programa que necesite un comportamiento impredecible.

Funciones clave:

    random.randint(a, b): Genera un número entero aleatorio entre 'a' y 'b' (incluyendo ambos).

    random.choice(lista): Elige un elemento al azar dentro de un grupo de opciones.

    random.random(): Genera un decimal aleatorio entre 0.0 y 1.0 (ideal para porcentajes).

```python
# Ejemplo 1: El dado de rol (Uso de randint)
dano_critico = random.randint(10, 50)
print(f"\n1. ¡Ataque exitoso! Has infligido {dano_critico} puntos de daño.")

# Ejemplo 2: Sistema de Botín o Loot (Uso de choice)
posibles_recompensas = ["Poción de Vida", "Mineral de Hierro", "Tun Tun Sahur", "Monedas de Oro"]
botin_obtenido = random.choice(posibles_recompensas)
print(f"2. Al abrir el cofre encontraste: {botin_obtenido}.")

# Ejemplo 3: Probabilidad de éxito (Uso de random)
probabilidad = random.random()
print(f"3. El número de la suerte fue {probabilidad}.")
```

### 3. Librería time (Controlando el Flujo)

La ejecución de un código en Python es casi instantánea. La librería time nos permite manipular el reloj interno del programa para crear pausas o medir cuánto tarda un proceso en ejecutarse.

Funciones clave:

    time.sleep(segundos): Pausa la ejecución del programa por la cantidad de segundos que le indiquemos.

    time.time(): Nos entrega la hora actual (en segundos transcurridos desde el 1 de enero de 1970, un estándar en informática).

```python
print("--- APLICACIONES DE TIME ---\n")

# Ejemplo 1: Efecto dramático / Simulador de carga (Uso de sleep)
print("1. Conectando con el servidor de mercado...")
time.sleep(2) # El programa se congela por 2 segundos
print("   ¡Conexión exitosa! Descargando precios...")
time.sleep(1.5) # Pausa de 1.5 segundos
print("   Datos actualizados.")

print("\n-----------------------------------\n")

# Ejemplo 2: Midiendo el rendimiento (Uso de time)
print("2. Midamos qué tan rápido eres.")
tiempo_inicio = time.time() # Guardamos la marca de tiempo inicial

# Le pedimos algo al usuario
input("Presiona ENTER lo más rápido que puedas...")

tiempo_final = time.time() # Guardamos la marca de tiempo final
tiempo_total = round(tiempo_final - tiempo_inicio) # Calculamos la diferencia

print(f"¡Te demoraste {tiempo_total} segundos en reaccionar!")
```
